```python
 import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/Human_body_weight'
```

```python
response = requests.get(url,headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36'})
```
